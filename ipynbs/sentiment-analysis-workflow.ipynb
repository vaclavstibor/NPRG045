## Named Entity Recognition

```python
from spacy import displacy

doc = nlp(article)

displacy.render(doc, style="ent", jupyter=True)
```

```python
from transformers import pipeline
from textblob import TextBlob

# Get only (ORG)ANIZATION entities
org_ents = [ent for ent in doc.ents if ent.label_ == 'ORG']

# Filter to unique organization entities
unique_entities = {ent.text: ent for ent in org_ents}
org_ents = list(unique_entities.values())

display(org_ents)
```

## Named Entity Sentiment Analysis

### 0. Approach

```python
entity_sentiments = []

for ent in org_ents:
    entity_sentiments.append((ent.text, ent.label_, ent.sentiment))

for entity, label, sentiment in entity_sentiments:
    display(f"Entity: {entity}, Label: {label}, Sentiment: {sentiment}")
```

### 1. Approach

```python
# For each entity, find sentences mentioning the entity and compute sentiment
for entity in org_ents:
    entity_sentences = [sentence.text for sentence in doc.sents if entity.text in sentence.text]
    entity_sentiment = [TextBlob(sentence).sentiment.polarity for sentence in entity_sentences]
    average_sentiment = sum(entity_sentiment) / len(entity_sentiment) if entity_sentiment else 0
    print(f"Entity: {entity.text}, Average Sentiment: {average_sentiment}")
```

### 2. Approach

```python
def get_entity_adjectives(entity):
    entity = "McDonald's Corp"  # Replace with your specific entity
    entity_adjectives = [token.text for token in doc if token.dep_ == 'amod' and token.head.text == entity]

    # Compute sentiment for each adjective
    sentiments = [TextBlob(adj).sentiment.polarity for adj in entity_adjectives]

    # Average sentiment for the entity
    average_sentiment = sum(sentiments) / len(sentiments) if sentiments else 0

    print(f"Entity: {entity}, Average Sentiment: {average_sentiment}")

for entity in org_ents:
    get_entity_adjectives(entity.text)
```
